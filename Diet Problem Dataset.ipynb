```python
m = len(nutrient_data)
n = len(nutrient_data.columns)

print("number of foods / drinks \t= ", m)
print("number of nutrition fields \t= ", n)
```

```python
#-----------------------------------
# Unpacking data to form .dat file
#-----------------------------------

def fix_string(val):
    vals = val.split(" ")
    return vals[0]

def fix_string_comma(val):
    vals = val.split(" ")
    return (vals[0] + ",")

food        = food_data.values
food_string = ""

for i in range(0, len(food)):
    food_string = fix_string((food[i])[0])
    print(food_string)
```

```python
#-----------------------------------
# Unpacking data to form .dat file
#-----------------------------------

def fix_food_string(val):
    vals = val.split(" ")
    return vals[0]

food        = food_data.values
costs       = cost_data.values
food_string = ""
cost_string = ""

for i in range(0, len(food)):
    food_string = fix_food_string((food[i])[0])
    cost_string = (costs[i])[0]
    print(food_string, cost_string)
```

```python
#-----------------------------------
# Unpacking data to form .dat file
#-----------------------------------


def specific_nutrient_string(vals, index):
    final_val = ""
    temp_val  = ""
    current   = ""
    
    current  = str(vals[index])
    temp_val = ""

    if ((current[:-1]).endswith("m")):
        temp_val = current[:-2]
    elif ((current).endswith("g")):
        temp_val = current[:-1]
    else:
        temp_val = current

    final_val = (final_val + temp_val + " ")
        
    final_val = final_val.strip()
    return final_val

def fix_nutrient_string(vals):
    final_val = ""
    temp_val  = ""
    current   = ""
    
    for val in vals:
        current  = str(val)
        temp_val = ""

        if ((current[:-1]).endswith("m")):
            temp_val = current[:-2]
        elif ((current).endswith("g")):
            temp_val = current[:-1]
        else:
            temp_val = current

        final_val = (final_val + temp_val + " ")
        
    final_val = final_val.strip()
    return final_val


food            = food_data.values
nutrients       = nutrient_data.values
food_string     = ""
nutrient_string = ""

for i in range(0, len(food)):
    food_string     = fix_food_string((food[i])[0])
    nutrient_string = fix_nutrient_string(nutrients[i])
    print(food_string, nutrient_string )
```
